```python
import sympy as sp
from sympy import sin, cos
```

```python
x = sp.Symbol('x')
expr = sin(x)**2 - cos(x)**2
expr
```

```python
# 1. определить корни
expr_solve = sp.solveset(expr)
expr_solve
```

```python
# 4. построить график
graph = sp.plot(expr)
```

```python
# 2. Найти интервалы, на которых функция возрастает
expr_diff = sp.diff(expr)
rise_expr = sp.solveset(expr_diff > 0)
rise_expr
```

```python
# 3. Найти интервалы, на которых функция убывает
down_expr = sp.solveset(expr_diff < 0)
down_expr
```

```python
# 5. Вычислить вершину
peak_expr = sp.solveset(expr_diff)
peak_expr
```

```python
# 6. Определить промежутки, на котором f > 0
peak_expr_origin = sp.solveset(expr > 0)
peak_expr_origin
```

```python
# 7. Определить промежутки, на котором f < 0
down_expr_origin = sp.solveset(expr < 0)
down_expr_origin
```
